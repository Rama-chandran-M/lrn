```python
## Basics

import numpy as np
arr1 = np.arange(15).reshape(3,5)
print(arr1)
print(arr1.shape)
print(arr1.ndim)
print(arr1.dtype)
print(arr1.dtype.name)
print(arr1.itemsize) ## size in byte of each element in the array
print(arr1.size) ## number of elements in the array
print(arr1.data)
print(type(arr1))
print(arr1.nbytes) ## 60 = 15 * 4
```

```python
## Array Creation

a = np.array([1,2,3])
print(a)
print(a.dtype)
b = np.array([1.2,2.3,3])
print(b)
print(b.dtype)

c = np.array([(1,2,3),(4,5,6)])
print(c)

d = np.array([[1,2],[3,4]],dtype=complex)
print(d)
```

```python
e = np.zeros((3,4))
print(e)
print(e.dtype)

f = np.ones((2,3),dtype=int)
print(f)
print(f.dtype)

g = np.ones((2,3,4))
print(g)
print(g.dtype) ## By default float 64

h = np.empty((2,3))
print(h)
print(h.dtype)
```

```python
i = np.arange(10,40,5)
print(i)
print(i.dtype)

j = np.arange(1,5,0.3)
print(j)

k = np.linspace(1,3,8) ## linspace receives the 3rd argument as number of elements instead of step, and produce evenly spaced numbers
print(k)

l = np.linspace(4,40) ## Generates 50 evenly spaced numbers Because the default value of num (number of samples) in np.linspace is 50.
print(l)
```

```python
arr = np.arange(12).reshape(3,4)
print(arr)

print(arr.sum(axis=0)) ## Sum of each column
print(arr.sum(axis=1)) ## Sum  of Each Row

print(arr.min(axis=1)) ## min in each row
print(arr.max(axis=0)) ##max in each column

print(arr.cumsum(axis=0)) ## cummulative sum along column
print(arr.cumsum(axis=1))
```

```python
## Universal Functions

a = np.arange(3)
print(a)
print(np.exp(a)) ## Produces e^x , where x is elements in the array
print(np.sqrt(a))
c = np.array([2.,-1.,4.])
print(np.add(a,c))
```

```python
## Indexing Slicing & Iterating

a = np.arange(10)**3
b = np.arange(10)**3
print(a)
print(a[3])
print(a[1:5])
print(a[:1]) # Prints elements upto index specified (excluding)
print(a[1:]) # Prints from the index specified
a[:6:2] = 1000 # Set every 2nd element to 1000 upto index specified (excluding)
print(a)
a=b # It is not creating a copy , it makes both variables point to what b points to
a[3::3] = 1000 # Set every 3nd index to 1000 from the index specified
print(a)
a[4:6] = 1000 # Set all index from 4 to index less than 6 as 1000
print(a)
print(a[::-1]) # Reverse
```

```python
def f(x,y):
    return 10*x+y

arr = np.fromfunction(f,(5,4),dtype=int) ## calls function f for each co-ordinates
print(arr)
print(arr[2,3])
print(arr[2][3])
print(arr[1:5]) ## Prints from 2nd to 5th row
print(arr[2:]) ##Prints from 3rd row
print(arr[:3]) ##Prints upto 3rd row
print(arr[0:5][1]) ## selects rows from 0 to 4 and then from those rows it chooses the 1st row
print(arr[0:5,1]) ## using tuple -> select row from 0 to 4 and then from it chooses the 1st column
print(arr[:,1]) ## same as previous one
print(arr[1:3,:]) ## select rows from 1,2 and then chooses all coumns in it
print(arr[3]) ##When fewer indices are provided than the number of axes, the missing indices are considered complete slices: equivalent to : arr[3,:]
print(arr[-1]) ## Print last row , equivalent to -> arr[-1,:]
print(arr[:,0:3:2]) ## selects all rows , startindex:endindex:step 
```

```python
a = np.array([[[0,1,2],[10,11,12]],[[100,101,102],[110,112,113]]])
print(a.ndim)
print(a.shape)

print(a[1])
print(a[1,:,:])
print(a[1,...])

print(a[0,:,2])

print(a[...,2]) ## Equivalent to a[:,:,2]

for r1 in a:
    for r2 in r1:
        print(r2)

for num in a.flat: ## flat attribute which is an iterator over all elements of the array
    print(num)
```
