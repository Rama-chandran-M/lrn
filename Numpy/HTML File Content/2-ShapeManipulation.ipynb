```python
## Changing the shape of an array
import numpy as np
rg = np.random.default_rng(1)
a = rg.integers(1,13,size=(3,4))
print(a)
print(a.shape)
```

```python
## the following three commands all return a modified array, but do not change the original array

print(a.ravel()) ## reval() -> Returns array element as flattened 
print(a)
print(a.reshape(4,3)) ## Returns with modified shape
print(a)
print(a.T) ## Returns Transpose
print(a)
```

```python
print(a)
print(a.resize(2,6)) ## Modifies the array itself with the size mentioned and returns 'None'.
print(a)

print(a.reshape(3,-1)) ##If a dimension is given as -1 in a reshaping operation, the other dimensions are automatically calculated
print(a)
print(a.reshape(-1))
print(a.reshape(12,-1))
```

```python
## Stacking together different Arrays
b = rg.integers(1,6,size=(2,2))
c = rg.integers(1,5,size=(2,2))
print(b)
print(c)
print(np.vstack((b,c)))
print(np.hstack((b,c)))
```

```python
rg = np.random.default_rng(1)
f = np.array([[1,2,3],[4,5,6]])
g = np.array([[7,8],[10,11]])
print(np.hstack((f,g))) ## Dimensions Conditions for hstack -> (n, a)  and  (n, b)  →  (n, a+b)

h = np.array([1,2,3])
i = np.array([[4,5,6],[7,8,9]])
print(np.vstack((h,i))) ## Dimensions Consitions for vstack -> (a, n)  and  (b, n)  →  (a+b, n)

m = rg.integers(1,10,size=(2,2,4))
n = rg.integers(1,10,size=(2,3,4))
print(np.hstack((m,n))) ## 3d hstack

o = rg.integers(1,10,size=(2,4,3))
p = rg.integers(1,10,size=(2,4,3))
print(np.vstack((o,p))) ## 3d vstack

## hstack,vstack are applicable to 4d,5d ... arrays , but only axis could have variable values
```

```python
d = np.array([1,2,3])
e = np.array([4,5,6])
## Column stack for 1d arrays
print(np.column_stack((d,e))) ## Converts 1D arrays into coulms of 2D array.

## column stack for 2d arrays works like hstack
## coulmn stack is not applicable for 3d,4d... kind of arrays
j = np.array([[1,2],[3,4]])
k = np.array([[5,6],[7,8]])
print(np.column_stack((j,k)))
l = np.array(([9,10],[11,12]))
print(np.column_stack((j,k,l)))
```

```python
## Splitting one array into several ones
rg = np.random.default_rng(1)
q = rg.integers(1,11,size=(2,12))
print(q)
print(np.hsplit(q,3)) ## splits the array into 3 equal parts, but only if it’s possible.

print(np.hsplit(q,(2,5))) ## splits the array at column indices 2 and 5.
```

```python
## np.array_split() splits an array into parts even when equal division is NOT possible.

r = np.arange(11)
print(r.size)
print(11//3)
print(np.array_split(r,3)) ## 11%3 -> rem = 2 , so 2 subarrays will have 1 element extra each than quotient.
```
