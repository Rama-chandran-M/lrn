```python
## Np array are faster and memory efficient than python
import sys
arr2 = range(1000)
print(sys.getsizeof(5)*len(arr2)) # A typical Python int takes 28 bytes on a 64-bit system.

arr3 = np.arange(1000)
print(arr3.itemsize*arr3.size) # On many systems, NumPy uses int32 → 4 bytes per element
```

```python
# -> Fast
import time
import numpy as np
size = 1000
l1 = range(size)
l2 = range(size)
start = time.time()
result1 = [(x+y) for x,y in zip(l1,l2)]
print("Python list took : ",(time.time()-start)*1000) # multiplying by 1000 changes seconds to milliseconds
n1 = np.arange(size)
n2 = np.arange(size)
start = time.time()
result2 = n1+n2
print("Numpy Array took : ",(time.time()-start)*1000)
```
