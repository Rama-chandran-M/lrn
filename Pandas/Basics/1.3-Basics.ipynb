```python
## Missing Data

import numpy as np
import pandas as pd
df=pd.DataFrame(
    {
        "A":np.array([1,2,3,4,5]),
        "B":np.array([34,21,6,5,76]),
        "C":np.array([99,44,22,54,12]),
        "D":np.array([11,22,33,44,55]),
        "E":np.array([54,43,32,21,10])
    }
)
print(df)

## Reindexing allows to change or add / delete index on a specified axis
df1 = df.reindex(index=([0,1,2,3]),columns=list(df.columns)+["F"])
print(df1)
df1.loc[[0,1],["F"]] = 1
print(df1)
```

```python
### dropna() -> drop rows that has missing values 

df1.dropna(how="any",inplace=True) ## This doesn't assign result so using inplace=True , it tells pandas to modify existing df itself ,without creating newone
print(df1)

## fillna() -> fills the missing values

df1 = df.reindex(columns=list(df.columns)+["F"]) ## Adding new column with df
print(df1)
df1.fillna(5,inplace=True)
print(df1)

df1["F"] = np.NaN
print(df1)

df1 = pd.isna(df1) ## Gets the boolean mask where the value is NaN
print(df1)
```

```python
## Operations

df3=pd.DataFrame(
    {
        "A":np.array([1,2,3,4,5]),
        "B":np.array([34,21,6,5,76]),
        "C":np.array([99,44,22,54,12]),
        "D":np.array([11,22,33,44,55]),
        "E":np.array([54,43,32,21,10])
    }
)
print(df3)
print(df3.mean())
print(df3.mean(axis=1))
```

```python
df3=pd.DataFrame(
    {
        "A":np.array([1,2,3,4,5]),
        "B":np.array([34,21,6,5,76]),
        "C":np.array([99,44,22,54,12]),
        "D":np.array([11,22,33,44,55]),
        "E":np.array([54,43,32,21,10])
    }
)
print(df3)
df4 = df3
s = pd.Series([1,3,5,np.nan,6,8]).shift(2) ## shift(2) -> moves data down by 2 rows
print(s)

## everything becomes NaN because since axis is not defined , so pandas tries to match s.index with df.columns , s.index != df.columns , so all become NaN
# df3 = df3.sub(s)
# print(df3) 

print(df3.sub(s,axis="index")) ##giving axis=index matches s index with df (rowwise)

## changing column to make column wise subtraction
s.index=["A","B","C","D","E","F"]
print(df3.sub(s,axis="columns"))

print(df3.sub(5)) ## applied to all values
print(df3["A"].div(7)) ## applied to column A
print(df3.iloc[4].mul(5)) ## applied to 5th row


# | Method       | Meaning        | Operator equivalent |
# | ------------ | -------------- | ------------------- |
# | `add()`      | Addition       | `+`                 |
# | `sub()`      | Subtraction    | `-`                 |
# | `mul()`      | Multiplication | `*`                 |
# | `div()`      | Division       | `/`                 |
# | `truediv()`  | True division  | `/`                 |
# | `floordiv()` | Floor division | `//`                |
# | `mod()`      | Modulus        | `%`                 |
# | `pow()`      | Power          | `**`                |
```

```python
## User defined Functions
print(df4)

## agg -> applies column-wise function , typically -> mean,sum,max
df5 = df4.agg(lambda x : np.mean(x) * 10) ## returns a series
print(df5)
print(df4.agg("mean")) ## returns a series of mean for each column
print(df4.agg(["sum","mean"])) ## returns a dataframe

## transform -> applies a function by default columnwise , preserves shape
print(df4.transform(lambda x : x*1000))
```

```python
## Merge - Concat

df = pd.DataFrame(np.random.randn(10,4))
# print(df)

p1 = df[:3]
p2 = df[3:7]
p3 = df[7:]
print(p2)
print(pd.concat([p1,p2,p3]))

# p2.drop(df.iloc[:,2],axis=1) -> df.iloc[:,2] returns data of column to be deleted , but drop removes with labels
p2 = p2.drop(p2.columns[2],axis=1)
print(pd.concat([p1,p2,p3]))

# To delete a row -> df.drop(2) (by label) , df.drop(df.index[2])(df.index[2] -> returns index label of 3rd row)
print(p2.index[2])
```

```python
## Join

left = pd.DataFrame({"key":["a","b"],"lval":[1,2]})
print(left)
right = pd.DataFrame({"key":["b","c"],"rval":[4,5]})
print(right)

print(pd.merge(left,right,on="key")) ## by default performs inner join

left = pd.DataFrame({"key":["a","a"],"lval":[1,2]})
print(left)
right = pd.DataFrame({"key":["a","c"],"rval":[4,5]})
print(right)
print(pd.merge(left,right,on="key")) ## 4 is repeated in rval column because left has duplicates "a"


left = pd.DataFrame({"key":["foo","foo"],"lval":[1,2]})
print(left)
right = pd.DataFrame({"key":["foo","foo"],"rval":[4,5]})
print(right)
print(pd.merge(left,right,on="key"))

left = pd.DataFrame({"key":["foo","bar"],"lval":[1,2]})
print(left)
right = pd.DataFrame({"key":["foo","bar"],"rval":[4,5]})
print(right)
print(pd.merge(left,right,on="key"))
```
