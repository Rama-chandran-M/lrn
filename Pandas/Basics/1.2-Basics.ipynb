```python
## Selecting the data
import numpy as np
import pandas as pd
df=pd.DataFrame(
    {
        "A":np.array([1,2,3,4,5]),
        "B":np.array([34,21,6,5,76]),
        "C":np.array([99,44,22,54,12]),
        "D":np.array([11,22,33,44,55]),
        "E":np.array([54,43,32,21,10])
    }
)
print(df)
## Get Item[]
print(df["A"])
print(df[1:3]) ## Slicing is always based on rows except by df is series
```

```python
## Selection by Label
print(df.loc[0])
print(df.loc[:,["A","C"]])
print(df.loc[1:3,["A","B","C"]]) ## end index also included
print(df.loc[0,"A"])
print(df.at[0,"A"]) ## Only for scalar , faster
```

```python
## Selection by Position -> 0th index based

print(df)
print(df.loc[3]) ## this is selection by label
print(df.iloc[3]) ## this is selection by posistion (3) in 0 based indexing
## print(df.iloc[0:3,["A":"C"]]) -> Not possible because "A" , "C" are not index works when 0,2 are given (same in next line)
print(df.iloc[:,[0,2]])
print(df.iloc[0:3,0:3])
print(df.iloc[1,1])
print(df.iat[1,1]) ## Used for scalar , fast
```

```python
## Boolean Indexing

print(df)
print(df[df["A"]>3]) ## Selects rws where df.A is greater than 3
print(df>35) ## Creates boolean dataframe
print(df[df>35]) ## Filter values based on the boolean mask
```

```python
df2 = df.copy()
df2["F"] = ["one","TWO","three","four","three"]
print(df2)
## df[] filter rows based on boolean series
## isin -> membership check , checks whether the value is present in the list
print(df2[df2["F"].isin(["two","four"])])
print(df2[df2["F"].str.lower().isin(["two","four"])])
```

```python
## Setting the values

s1 = pd.Series([1,2,3,4,5])
df2["G"] = s1
print(df2)

## Setting values by label
df2.at[1,"B"] = 5 ## can use .loc
print(df2)

## Setting values by position
df2.iat[2,1] = 999 ## can use .iloc
print(df2)

## Setting by assigning with a numpy array
df2.loc[:,"D"] = np.array([3]*len(df2))
print(df2)

df2 = df2.drop("F",axis=1) ## deleting a column

df3 = df2.copy()
df3[df3 > 0] = -df3 ##this -df3 can be done only if all values are negative , so removed column F in previous step
print(df3)
```
