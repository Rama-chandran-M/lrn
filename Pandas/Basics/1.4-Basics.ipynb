```python
import pandas as pd
import numpy as np

## Grouping
df = pd.DataFrame(
    {
        "A":["foo","bar","foo","bar","foo","bar","foo","bar"],
        "B":["one","one","two","three","two","two","one","three"],
        "C":np.random.randn(8),
        "D":np.random.randn(8)
    }
)
print(df)
print(df.groupby("A").sum())
print(df.groupby("A")[["C","D"]].sum())
print(df.groupby(["A","B"]).sum())
```

```python
## MultiIndex Concept

## Single Index
df = pd.DataFrame(
    {"marks":[90,95,99]},
    index=["Alice","Bob","Charlie"]
)
print(df)

## MultiIndex -> hierarchial index
array = [
    ["CSE","CSE","IT"],
    ["Alice","Bob","Charlie"]
]
index = pd.MultiIndex.from_arrays(array,names=["DEPT","Student"])
df = pd.DataFrame(
    {"mark":[90,95,99]},
    index=index
)
print(df)
```

```python
## Reshaping
arrays =[
    ["bar","bar","baz","baz","foo","foo","qux","qux"],
    ["one","two","one","two","one","two","one","two"]
]
index = pd.MultiIndex.from_arrays(arrays,names=["first","second"]) ## for index -> names will split across index section in columns based on array
df = pd.DataFrame(
    np.random.randn(8,2),
    index= index,
    columns = ["A","B"]
)
print(df)

## Multi Index Columns
array =[["Math","Math","Science","Science"],["Mid","Final","Mid","Final"]]
columns = pd.MultiIndex.from_arrays(array,names=["Subject","Exam"]) ## for columns -> names will split into rows as a collective label for column label
df = pd.DataFrame(np.random.randn(4,4),columns=columns,index=["a","b","c","d"])
print(df)

## instead of repeating in arrays we can use -> from_prodct
cols = pd.MultiIndex.from_product([["Math","Science"],["Mid","Final"]])
df = pd.DataFrame(np.random.randn(5,4),columns=cols)
print(df)
```

```python
## Stack() -> rotates column to rows , moves the columns labels into a new inner index level
df2=pd.DataFrame(np.random.randn(4,2),columns=["A","B"])
print(df2)
df2 = df2.stack(future_stack=True)
print(df2)

## unstack()
print(df2.unstack()) ## unstack last level if index
print(df2.unstack(0)) ## unstack 0th level of index (from left)
```

```python
## Pivot Tables

df = pd.DataFrame(
    {
        "A":["one","one","two","three"]*3,
        "B":["A","B","C"]*4,
        "C":["foo","foo","foo","bar","bar","bar"]*2,
        "D":np.random.randn(12),
        "E":np.random.randn(12)
    }
)
print(df)
pt = pd.pivot_table(df,values="D",index=["A","B"],columns=["C"]) ## default aggregation function mean is appied
print(pt)

df.loc[:,["B"]] = ["A","A","C"]*4
print(df)
print(pd.pivot_table(df,values=["D"],index=["A","B"],columns=["C"],aggfunc="sum"))
```

```python
## Time Series , abbrevations -> rng - range ,ts -timeseries
rng = pd.date_range("1/1/2025",periods=100,freq='s')
ts = pd.Series(np.random.randint(0,501,len(rng)),index=rng)
print(ts)
print(ts.resample("1Min").count())
print(ts.resample("5Min").sum())
```

```python
## series.tz_localize()

rng = pd.date_range("3/6/2012",periods=5,freq='D')
ts = pd.Series(np.random.randint(0,101,len(rng)),rng)
print(ts)
ts_utc = ts.tz_localize("UTC") ## attaches timezone to timestamp , UTC stands for Coordinated Universal Time
print(ts_utc)

## series.tz_converts() -> converts timezone to another timezone -> Only timezone info (+00:00) was added
converted = ts_utc.tz_convert("Asia/Kolkata")## ->   +05:30
print(converted)
print(ts_utc.tz_convert("US/Eastern")) ## ->   -05:00
```

```python
rng = pd.date_range("3/6/2012",periods=5,freq='D')
print(rng)
print(rng+pd.offsets.BusinessDay(5)) ## adds 5 bussiness days to each date , bussiness day = Monday to friday
print(rng+pd.offsets.MonthEnd()) ## Moves to Month end
print(rng+pd.offsets.MonthEnd(5)) ## Moves to next 5th month end
```
