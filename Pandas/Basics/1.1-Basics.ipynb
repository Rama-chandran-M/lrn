```python
import pandas as pd
import numpy as np

## Oject Creation
## Series -> 
a = pd.Series([1,2,3,np.nan,4,5])
print(a) 
```

```python
## Dates

dates = pd.date_range("20250101",periods=6) ## periods -> Number of dates to generate , default frequency is day
print(dates)

dates2 = pd.date_range("20240101",periods=2,freq="ME") ## ME -> Month End
print(dates2)

dates3 = pd.date_range("20240101",periods=2,freq="h") ## h -> hours
print(dates3)

dates4 = pd.date_range("20250101",periods=2,freq="W") ## W -> Weekly (Sunday)
print(dates4)
```

```python
b = pd.DataFrame(np.random.randint(1,11,size=6))
print(b)
```

```python
date = pd.date_range("20250101",periods=6)
## np.random.randn(n) generates n numbers whose mean is 0(if n is large) , std is 1 .
c = pd.DataFrame(np.random.randn(6,4),index=date,columns=list("ABCD"))
print(c)
```

```python
## Creating a DataFrame by passing a dictionary of objects where the keys are the column labels and the values are the column values.

## Each key → column name
## Each value → column data (object)

df = pd.DataFrame(
    {
        "A":np.array([1,2,3,3,5]),
        "B":pd.Timestamp("20240202"), ## data-time object
        "C":pd.Series(1,index=list(range(5))),
        "D":np.array([3]*5,dtype="int32"),
        "E":pd.Categorical(["test","train","test","train","test"]),
        "F":"foo",
        "G":np.array([4,7,5,3,9]),
    }
)
print(df)

## here the values are scalar(single value) or list or array , if list or array is used then the length of all used list and arrray
## must be same

df2 = pd.DataFrame(
    {
        "A":np.array([1]),
        "B":pd.Series(2),
        "C":"boo"
    }
)
print(df2)
```

```python
## NumPy arrays have one dtype for the entire array while pandas DataFrames have one dtype per column

print(df.dtypes)
print(df2.dtypes)
```

```python
## Viewing the Data
print(df.head())
print(df.tail(3))
print(df.index)
print(df.columns)
print(df.to_numpy()) ## Returns a numpy respresentation of the data
print(df.describe()) ## shows a quick statistic summary of your data
print(df.T) ## Transpose of data
print(df.sort_index(ascending=False))
print(df.sort_index(axis=1,ascending = False)) ##axis=1 - > horizontal , axis=0 -> vertical
df_sorted = df.sort_index(axis=0, ascending=False).sort_index(axis=1, ascending=False) ## rows and column indices are sorted independently
print(df_sorted)
print(df)
print(df.sort_values(by='A',ascending=False)) ## sort the rows based on column 'A'
print(df.sort_values(by=['A','G'],ascending=[False,True])) ## sort rows based on multiple columns , useful when 1st column has same values
print(df.index)
df3=pd.DataFrame(
    {
        "A":np.array([1,2,3]),
        "B":np.array([5,6,7]),
        "C":np.array([3,2,1]),
    }
)
print(df3)
print(df3.sort_values(by=1,axis=1,ascending=False))## that row should have same dtype
```
